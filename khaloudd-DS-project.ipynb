10-To query records in a dataframe under certain conditions, the Pandas library introduces the query function of the dataframe, a function that makes it easy for you to perform queries with parameters and conditions.
in my case ,I want to inquire about the Male and Female students from the State of Kuwait who have raised the hand more than 50 times in the class and so that we display the gender property and the number of times the hand was raised only,I use the query function in the following way.

```python
df.query("gender == 'F' and Topic == 'IT' and raisedhands > 50" )[["gender","raisedhands"]]
```

```python
df.query("gender == 'M' and Topic == 'IT' and raisedhands > 50" )[["gender","raisedhands"]]
```

11-In many experiments, we need to obtain a graphical figure showing the number of recurrences of records for a particular property (nominal or ordinal) according to the range of values ​​it contains. The Seaborn library provides a countplot function that does this. For example, if we want to know the numbers of students by nationalities, we do the following.

```python
fig = sns.countplot(x="NationalITy", data=df, palette="muted");
fig.set_xticklabels( fig.get_xticklabels() ,  rotation=70)
plt.show()
```

12-I call the countplot function and we pass it first the column we want to count based on its value which is in our case NationalITy and then we select the data frame we are working on, and finally choose the color set of the shape. The function set_xticklabels we use to rotate (70 degrees) the names of nationalities so that they do not overlap between them and the figure appears in an ordered manner.
Through the previous graph, and a quick look at it, we get information that most of the students are from Kuwait and Jordan. The following example shows how to do the same for the student's score property

```python
fig = sns.countplot(x="Class", data=df, palette="muted");
plt.show()
```

I notice from the figure that most of the students got an average grade of M in the subjects in general, and the number of students who got a high grade of H is slightly more than the students who got a low grade of L.

13-Displaying the numbers of students by subjects with the breakdown of numbers by degree We pass the hue parameter countplot function and specify the name of the degree property as follows

```python
plt.subplots(figsize=(13,10))
fig = sns.countplot(x="Topic", data=df, palette="Set2",hue="Class");
fig.set_xticklabels( fig.get_xticklabels() ,  rotation=70)
plt.show()
```

14-The graphs may sometimes be not detailed, so we can use the so-called swarm shapes so that each point that represents a record in the data is plotted as:

```python
Raised_hand = sns.swarmplot(x="Class", y="raisedhands", data=df,  palette="muted")
plt.show()
```

15-From the previous figure, it is clear to us that most of the students who raised their hand in class more than 60 times got a high grade of H, while most of the students who got a low grade L did not raise their hand more than 20 times in class, while the students who got an average grade were distributed Hand raise times they have on the property area.
The Seaborn library provides another function called joinplot to plot a graph showing the relationship of two variables to each other.

```python
fig = sns.jointplot(x="raisedhands", y="VisITedResources", data=df, kind="reg", size=6);
plt.show()
```
